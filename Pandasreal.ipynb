##Đọc dữ liệu

Chúng ta có thể đọc dữ liệu từ các nguồn khác nhau như CSV, Excel, SQL databases, JSON, v.v.

Chúng ta dùng phương thức: pd.read_csv(), pd.read_excel(), pd.read_sql(), pd.read_json():

read_csv:
+ A simple way to store big data sets is to use CSV files (comma separated files).

+ CSV files contains plain text and is a well know format that can be read by everyone including Pandas.

    + pd.read_csv('filepath_or_buffer', sep, header, names, index_col, dtype)

    +filepath_or_buffer (str hoặc path-like): Đường dẫn hoặc URL của tệp cần đọc.

    + sep (str, default=','): Dấu phân cách giữa các trường dữ liệu trong tệp CSV. Mặc định là dấu ','.

    + header (int, list of int, default='infer'): Số dòng ở đầu tệp để sử dụng làm header. Nếu 'infer', Pandas tự động xác định dòng header.(Bằng cách lấy dòng đầu tiên của file )

    + names (array-like, default=None): Danh sách tên cột. Nếu không được cung cấp và header=None, tên cột sẽ được tạo theo dạng '0', '1', v.v.

    + index_col (int, str, sequence[int/str], or False, default=None(không sử dụng cột nào làm chỉ số, tự tạo)): Cột được chọn làm chỉ số (index). Mặc định là None, sử dụng chỉ số mặc định của Pandas.

    + dtype (Type name or dict of column -> type, default=None (Pandas sẽ tự động xác định kiểu dữ liệu của mỗi cột)): Kiểu dữ liệu của cột hoặc dữ liệu. Có thể là một kiểu dữ liệu hoặc một từ điển chỉ định kiểu cho từng cột.

##Xóa cột hoặc hàng từ DataFrame

+ Chúng ta có thể dùng method drop:

+ DataFrame.drop(labels=None, axis=0, index=None, columns=None, inplace=False, errors='raise')

    + labels: Tên hoặc chỉ số của cột hoặc hàng cần xóa. Đây có thể là một chuỗi hoặc danh sách.

    + axis: Xác định xem bạn muốn xóa theo chiều ngang (1) hay chiều dọc (0). Mặc định là 0 (xóa hàng).

    + index: Tương đương với labels khi axis=0. Là chỉ số hoặc danh sách chỉ số của hàng cần xóa.

    + columns: Tương đương với labels khi axis=1. Là tên hoặc danh sách tên của cột cần xóa.

    + inplace: Nếu True, thay đổi sẽ được thực hiện trực tiếp trên DataFrame và không trả về DataFrame mới. Nếu False (mặc định), một DataFrame mới sẽ được trả về và DataFrame gốc sẽ không bị thay đổi.

    + errors: Xác định cách xử lý nếu labels không được tìm thấy. Giá trị mặc định là 'raise' để bật ra một ngoại lệ, nhưng bạn có thể đặt thành 'raise', 'coerce', hoặc 'ignore'. 'raise' sẽ ném một ngoại lệ nếu không tìm thấy, 'coerce' sẽ chuyển đổi labels thành không có lỗi, và 'ignore' sẽ bỏ qua lỗi.

```python
# Tạo một DataFrame ví dụ
data = {'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]}
df = pd.DataFrame(data)

# Xóa cột 'B' và 'C'
df.drop(columns=['B', 'C'], inplace=True)

# Xóa hàng có chỉ số 0
df.drop(index=0, inplace=True)

# Hiển thị DataFrame kết quả
print(df)
```

## Điền giá trị cho dữ liệu thiếu.


+ Chúng ta có thể dùng method fillna()
+ DataFrame.fillna(value=None, method=None, axis=None, inplace=False, limit=None, downcast=None)

    + value: Giá trị bạn muốn điền vào ô dữ liệu thiếu. Điều này có thể là một giá trị cụ thể, một chuỗi, hoặc một giá trị được tính toán.

    + method: Các phương thức để điền giá trị. Một số giá trị có thể là: 'ffill' hoặc 'pad': Sử dụng giá trị từ hàng trước đó để điền vào ô thiếu 'bfill' hoặc 'backfill': Sử dụng giá trị từ hàng sau đó để điền vào ô thiếu

    + limit: Số lượng giá trị cần điền (tính từ đầu hoặc từ cuối tùy thuộc vào giá trị của method).

    + downcast: Chỉ định cách giảm độ chính xác của dữ liệu (nếu có). Các giá trị có thể là 'integer', 'signed', 'unsigned', 'float'.


```python
import numpy as np

# Tạo một DataFrame ví dụ với dữ liệu thiếu
data = {'A': [1, np.nan, 3], 'B': [4, 5, np.nan], 'C': [7, 8, 9]}
df = pd.DataFrame(data)

# Điền giá trị 0 vào các ô dữ liệu thiếu
df.fillna(value=0, inplace=True)

# Hiển thị DataFrame kết quả
print(df)
```

```python
#Ví dụ để hiểu method
data = {'A': [1, np.nan, 3, np.nan, 5],
        'B': [np.nan, 2, 3, 4, np.nan],
        'C': [1, 2, np.nan, 4, 5]}
df = pd.DataFrame(data)

print("Before: \n", df)
# Sử dụng forward fill để điền giá trị từ hàng trước đó
df_ffill = df.fillna(method='ffill')
print("DataFrame after forward fill:")
print(df_ffill)

# Sử dụng backward fill để điền giá trị từ hàng sau đó
df_bfill = df.fillna(method='bfill')
print("\nDataFrame after backward fill:")
print(df_bfill)
```

```python
#Ví dụ về limit
data = {'A': [1, np.nan, np.nan, 4, 5],
        'B': [np.nan, 2, 3, 4, 5],
        'C': [1, 2, np.nan, 4, 5]}
df = pd.DataFrame(data)

# Sử dụng forward fill với limit=1
df_ffill_limit = df.fillna(method='ffill', limit=1)
print(df_ffill_limit)
```

##Tổ chức và thống kê dữ liệu theo nhóm.


+ Chúng ta có thể dùng df.groupby(), df.pivot_table():
+ Ở đây chúng tớ chỉ giới thiệu groupby()
+ DataFrame.groupby(by=None, axis=0, level=None, as_index=True, sort=True, group_keys=True, squeeze=False, observed=False, **kwargs)

    + by: Cột hoặc danh sách các cột dựa trên đó bạn muốn nhóm dữ liệu. Đây có thể là tên cột hoặc các chuỗi đặc tả điều kiện nhóm.

    + level: Cho MultiIndex DataFrame, chỉ định cấp độ cần thực hiện groupby.

    + as_index: Nếu True (mặc định), cột hay cột cấp độ sẽ trở thành index của DataFrame kết quả. Nếu False, index mới sẽ được tạo ra.

    + sort: Nếu True (mặc định), kết quả sẽ được sắp xếp theo giá trị tăng dần của các cột nhóm. Nếu False, không sắp xếp.

    + group_keys: Nếu True (mặc định), keys của nhóm sẽ được hiển thị trong kết quả. Nếu False, keys sẽ không được hiển thị.

    + squeeze: Nếu True, nếu chỉ có một cột nhóm, kết quả sẽ được squeeze thành Series. Nếu False (mặc định), kết quả là DataFrame.

    + observed: Chỉ định cách xử lý các giá trị không quan sát được trong các biến nhóm. Nếu True, sẽ thực hiện groupby dựa trên tất cả các giá trị quan sát được. Nếu False (mặc định), sẽ chỉ thực hiện groupby dựa trên các giá trị xuất hiện trong dữ liệu.

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Value': [1, 2, 3, 4, 5, 6],
    'Score': [10, 20, 30, 40, 50, 60]
}

df = pd.DataFrame(data)

# Tham số by
grouped_by_category = df.groupby(by='Category')
print("Grouped by Category:")
print(*grouped_by_category)
print(grouped_by_category.sum())

# Tham số as_index
grouped_as_index = df.groupby(by='Category', as_index=False)
print("\nGrouped with as_index=False:")
print(grouped_as_index.sum())

#Tham số group_keys
grouped_no_keys = df.groupby(by='Category', group_keys=False)
print("\nGrouped with group_keys=False:")
print(grouped_no_keys.sum())
```

```python
# Tham số observed
data = {
    'Category': ['A', 'B', 'A', 'B', np.nan, 'B'],
    'Value': [1, 2, 3, 4, 5, 6],
    'Score': [10, 20, 30, 40, 50, 60]
}

df = pd.DataFrame(data)

# Kết quả nhóm sẽ loại bỏ nhóm có giá trị NaN khi observed=False.
grouped_observed_false = df.groupby(by='Category', observed=False)
print(grouped_observed_false)
print("Grouped with observed=False:")
print(*grouped_observed_false)

# Kết quả nhóm sẽ bảo toàn nhóm có giá trị NaN khi observed=True.
grouped_observed_true = df.groupby(by='Category', observed=True)
print("\nGrouped with observed=True:")
print(*grouped_observed_true)
```

##Kết hợp các DataFrame dựa trên cột chung hoặc nối các DataFrame theo chiều dọc hoặc ngang.

+ Chúng ta có thể dùng method merge() hoặc concat()
+ Ở đây chúng tớ chỉ giỏi thiệu merge()
+ pandas.DataFrame.merge(right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, sort=False, suffixes=('_x', '_y'), copy=True, indicator=False, validate=None)

    + right: DataFrame sẽ được kết hợp với DataFrame gọi phương thức (self).

    + how: Loại kết hợp. Các giá trị có thể là:
    
    'inner': Kết hợp trong, chỉ giữ các hàng có giá trị khớp trong cả hai DataFrame (mặc định).
    'outer': Kết hợp ngoại, giữ tất cả các hàng từ cả hai DataFrame và điền giá trị NaN nếu không có sự khớp.
    'left': Kết hợp trái, giữ tất cả các hàng từ DataFrame gọi phương thức và điền giá trị NaN nếu không có sự khớp trong DataFrame bên phải.
    'right': Kết hợp phải, giữ tất cả các hàng từ DataFrame bên phải và điền giá trị NaN nếu không có sự khớp trong DataFrame gọi phương thức.
    
    + on: Tên hoặc danh sách tên cột chung giữa hai DataFrame.

    + left_on và right_on: Tên hoặc danh sách tên cột chung trong DataFrame gọi phương thức và DataFrame được kết hợp, nếu các cột không có cùng tên. -> Dùng để đặt tên các cột mà chung muốn khác tên.

    + left_index và right_index: Nếu True, sử dụng index của DataFrame gọi phương thức hoặc DataFrame được kết hợp làm cột kết hợp.

    + sort: Nếu True (mặc định), sắp xếp kết quả theo cột kết hợp.

    + suffixes: Tuple của hai chuỗi, đặc tả hậu tố của các cột có tên trùng nhau từ hai DataFrame.

    + copy: Nếu True (mặc định), tạo một bản sao của DataFrame để tránh ảnh hưởng đến DataFrame gốc.

    + indicator: Nếu True, thêm một cột '_merge' cho biết hàng đến từ DataFrame nào (left_only, right_only, hoặc both).

    + validate: Kiểm tra loại kết hợp. Các giá trị có thể là None (mặc định), 'one_to_one', 'one_to_many', 'many_to_one', 'many_to_many'.

```python
# Tạo hai DataFrame ví dụ
df1 = pd.DataFrame({'key': ['A', 'B', 'C'], 'value': [1, 2, 3]})
df2 = pd.DataFrame({'key': ['A', 'B', 'D'], 'value': [4, 5, 6]})

# Merge theo cột 'key' với kiểu kết hợp 'inner'
merged_df = pd.merge(df1, df2, on='key', how='inner')

# Hiển thị DataFrame kết quả
print(merged_df)
```

```python
#Ví dụ về right, how, và on:
# Tạo hai DataFrame
df1 = pd.DataFrame({'key': ['A', 'B', 'C'], 'value': [1, 2, 3]})
df2 = pd.DataFrame({'key': ['A', 'B', 'D'], 'value': [4, 5, 6]})

# Merge DataFrame sử dụng 'key' làm cột kết hợp, và giữ tất cả các hàng từ cả hai DataFrame.
result_inner = df1.merge(df2, how='inner', on='key')
result_outer = df1.merge(df2, how='outer', on='key')
result_left = df1.merge(df2, how='left', on='key')
result_right = df1.merge(df2, how='right', on='key')

print("Inner Merge:")
print(result_inner)

print("\nOuter Merge:")
print(result_outer)

print("\nLeft Merge:")
print(result_left)

print("\nRight Merge:")
print(result_right)
```

```python
#Ví dụ về left_on và right_on
# Tạo hai DataFrame với các cột khác tên
df1 = pd.DataFrame({'key_left': ['A', 'B', 'C'], 'value': [1, 2, 3]})
df2 = pd.DataFrame({'key_right': ['A', 'B', 'D'], 'value': [4, 5, 6]})

# Merge DataFrame sử dụng 'key_left' và 'key_right' làm cột kết hợp.
result = df1.merge(df2, how='inner', left_on='key_left', right_on='key_right')

print("Merge with left_on and right_on:")
print(result)
```

```python
#Ví dụ về left_index và right_index
# Tạo hai DataFrame với index khác nhau
df1 = pd.DataFrame({'value': [1, 2, 3]}, index=['A', 'B', 'C'])
df2 = pd.DataFrame({'value': [4, 5, 6]}, index=['A', 'B', 'D'])

# Merge DataFrame sử dụng index của cả hai DataFrame làm cột kết hợp.
result = df1.merge(df2, how='inner', left_index=True, right_index=True)

print("Merge with left_index and right_index:")
print(result)
```

```python
#Ví dụ về suffixes
df1 = pd.DataFrame({'key': [1, 2, 3], 'value': ['A', 'B', 'C']})
df2 = pd.DataFrame({'key': [1, 2, 4], 'value': ['X', 'Y', 'Z']})

# Merge với suffixes để đặt hậu tố cho các cột trùng tên
result_suffixes1 = df1.merge(df2, how='outer', on='key')
result_suffixes = df1.merge(df2, how='outer', on='key', suffixes=('_df1', '_df2'))
print("Merge with suffixes:")
print(result_suffixes1)
print(result_suffixes)
```

```python
#Ví dụ về Indicator
df1 = pd.DataFrame({'key': [1, 2, 3], 'value': ['A', 'B', 'C']})
df2 = pd.DataFrame({'key': [1, 2, 4], 'value': ['X', 'Y', 'Z']})

# Merge với indicator=True để thêm cột '_merge'
result_indicator = df1.merge(df2, how='outer', on='key', indicator=True)
print("Merge with indicator:")
print(result_indicator)
```

```python
#ví dụ về validate
df1 = pd.DataFrame({'key': [1, 2, 3], 'value': ['A', 'B', 'C']})
df2 = pd.DataFrame({'key': [1, 2, 2], 'value': ['X', 'Y', 'Z']})

# Merge với validate='one_to_one', nhưng có cột key không đảm bảo one-to-one
result_validate_error = df1.merge(df2, how='outer', on='key', validate='one_to_one')
print("Merge with validate='one_to_one' (Error):")
print(result_validate_error)
```

#8.Thống kê và tổng hợp

### GroupBy

groupby là câu lệnh cho phép bạn áp dụng những hàm số trên measurements dựa trên việc phân nhóm dữ liệu theo các dimensions.
=> tóm lại là groupby sẽ là việc với 2 biến là dimension và mesurements.
-  khái niệm measurement và dimension :
  - measurement là những biến có thể cộng trừ nhân chia và đo đếm được
  - dimension là những biến dùng để phân nhóm dữ liệu.
Ví dụ  ở trên Total cases là một measurement có thể đo theo dimension là Crime gồm các loại tội  .

Cú pháp của hàm **df.groupby()** khá đơn giản:
-  <Tên dataframe>.groupby('dimension')['meansurement'].(hàm tính toán)

```python
data.groupby('Location')['Total cases'].mean()
```

```python
result = data.groupby('Crime')[['Total cases','Year']].mean()
result
```

Theo cách trên thì ta chỉ áp dụng được với những hàm tính toán như sum, avg, min, max có sẵn trong dataframe. Nếu muốn sử dụng groupby() cho mọi biến đổi hàm chúng ta có thể dùng hàm lambda trong apply().



```python
def nhandoi(x):
  return x*2

result = data.groupby('Crime')['Total cases'].apply(lambda x : nhandoi(x))
```

### Pivot_table

Bạn sẽ hình dung ra cách áp dụng của pivot_table() thông qua ví dụ bên dưới.

```python
import numpy as np

pd.pivot_table(data,
               columns = ['Crime'],
               values = 'Total cases',
               aggfunc = {
                  np.sum
               })
```

Dưới đây là một số tham số quan trọng của hàm pivot_table:

- data: Đối số bắt buộc, là DataFrame chứa dữ liệu bạn muốn sử dụng để tạo pivot table.

- values: Đối số bắt buộc, chỉ định cột hoặc danh sách các cột chứa giá trị mà bạn muốn tổng hợp trong pivot table.

- index: Các cột trong DataFrame được sử dụng làm chỉ số (index) của pivot table. Các giá trị trong các cột này sẽ trở thành các dòng (rows) của pivot table.

- columns: Các cột trong DataFrame được sử dụng làm các cột của pivot table.

- aggfunc: Hàm tổng hợp (aggregation function) được áp dụng cho các giá trị được chỉ định bởi values. Mặc định là 'mean', nhưng bạn có thể sử dụng các hàm tổng hợp khác như 'sum', 'count', 'min', 'max', vv.

- fill_value: Giá trị được sử dụng để thay thế các giá trị thiếu (NaN).

- margins: Nếu True, thêm một dòng và một cột tổng hợp (tổng cộng) vào pivot table.

- margins_name: Tên của dòng và cột tổng hợp, nếu margins là True.

- dropna: Nếu True, loại bỏ các dòng có giá trị NaN.

**Làm sao để qui định mỗi measurement một công thức?**



```python
pd.pivot_table(data,
               columns = ['Crime'],
               values = ['Total cases','Year'] ,
               aggfunc = {
                  'Total cases': np.sum,
                  'Year': np.mean
               })
```

  - Pivot_table : kết quả trả về là một đối tượng có kiểu **DataFrame** ***mới***

  - **pivot_table**: thường được sử dụng khi bạn muốn biến đổi dữ liệu và tạo ra một bảng tổng hợp, có thể thực hiện cả hai bước phân nhóm và tổng hợp một cách thuận tiện.
