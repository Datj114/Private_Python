**Một số hàm hay để hiểu về python:**



*   dir(): trả về danh sách thuộc tính và phương thức của một đối tượng
Ví dụ: dir(str)

*  help(): Lệnh này được sử dụng để xem tài liệu về một đối tượng, một lớp, một phương thức hoặc một module.
Ví dụ: help(str)


*   type(): Lệnh này trả về kiểu của một đối tượng.
Ví dụ: type(“hit”)


*   id(): (python không hỗ trợ lấy giá trị trực tiếp địa chỉ của biến mà nó sử dụng mã định danh) lệnh này trả về định danh duy nhất của một đối tượng.
Ví dụ: id(a)

*  isinstance(): Lệnh này kiểm tra xem một đối tượng có phải là một thể hiện của một lớp cụ thể hay không.
Ví dụ: isintance(“ahahha”, string)



*   globals() và locals(): Các lệnh này trả về danh sách các biến toàn cục và cục bộ.

*   callable(): Lệnh này kiểm tra xem một đối tượng có thể gọi được hay không (nghĩa là nó có phải là một hàm hay không).


*   getattr(), setattr(), và delattr(): Các lệnh này cho phép bạn lấy, thiết lập và xóa các thuộc tính của một đối tượng.

*   importlib.reload(): Lệnh này cho phép bạn tải lại một module đã được import.


*  name và main: Biến __name__ cho biết tên của module hiện tại, trong khi __main__ chỉ ra rằng module hiện tại đang được chạy như là chương trình chính.
