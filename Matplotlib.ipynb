#4, Matplotlib Line

- Hiểu đơn giản là kiểu của đường sẽ được vẽ.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 6, 8])
y = np.array([3, 8, 1, 10])

plt.plot(x, y, marker = 'o', linestyle = ':', color = 'hotpink', linewidth = 1.9)
plt.plot(y)
plt.show()
```

- Bạn có thể viết là linestyle = ':' hay ls = ':' đều được.
- Hay thay ':' bằng 'dotted'.
- Tùy thuộc theo sở thích của bạn:

- + 'solid' (default)	'-'
- + 'dotted'	':'
- + 'dashed'	'--'
- + 'dashdot'	'-.'
- + 'None'	"" or ''

- Như linestyle thì color cũng được viết tắt là c. Và vẫn tùy thuộc vào bạn :D

- + '#FF0000', '#000000', '#0000ff', ... (color hex)

#5, Matplotlib Label, Title, Grid, Legend (**Tôi** gọi chung là trang trí)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 5, 1, 6, -5, 3])
y = np.array([5, 5, 2.3, 3, 0, 0])

font1 = {'family':'serif', 'size':30, 'c': 'b'}

plt.plot(x, y, label = "Ten cac doan cung mau so 1")
plt.plot(x, label = "Ten cac doan cung mau so 2")
plt.plot(y, label = "Ten cac doan cung mau so 3")

plt.xlabel("Truc Ox: Nam ngang", loc = 'right', fontdict = font1)
plt.ylabel("Truc Oy: Nam doc", loc = 'bottom', fontdict = font1.update({'size': 10}) != None or font1)
plt.title("Tieu de: O tren", loc = 'left', fontdict = font1.update({'c': 'r'}) != None or font1)

plt.grid(axis= 'x', c = 'pink', ls = ':', lw = 2)

plt.legend(loc = "upper left")

plt.show()
```

# 6, Matploylib Bars

**Matblotlib Bars giúp biểu diễn, trực quan hoá dữ liệu dưới dạng cột, thuận lợi cho việc so sánh, sự biến động của giá trị**

- Hàm bar() có chức năng vẽ biểu đồ thanh dọc
-Hàm bar() lấy các đối số (arguments) mô tả bố cục của các thanh. Các danh mục và giá trị của chúng được biểu thị bằng dạng mảng thứ nhất (x) và dạng mảng thứ hai (y)

```python
import matplotlib.pyplot as plt
import numpy as np
# mảng danh mục
x = np.array(["A", "B", "C", "D"])
# mảng giá trị
y = np.array([3, 8, 1, 10])
plt.bar(x,y)
plt.show
```

**Một số parameter**

plt.bar(x,y, width=0.8, bottom=None, align='center', data=None, **kwargs)
x: Dãy giá trị trục x.
-height: Dãy giá trị chiều cao của các thanh.
 - width: Độ rộng của các thanh (mặc định là 0.8).
 - bottom: Dãy giá trị hoặc số liệu để xác định bottom của các thanh (mặc định là None).
 - align: Căn chỉnh các thanh so với vị trí x ('center', 'edge').
- data: Dữ liệu DataFrame (mặc định là None).
-  **kwargs: Các tham số khác như màu sắc, độ trong suốt, v.v.

```python
# sử dụng từ khoá color để thay đổi màu sắc của bars
plt.bar(x,y,color= 'orange')
```

# 7, Matplotlib Pie Charts

**Hàm pie() giúp vẽ biểu đồ hình tròn**

- kích thước mỗi phần = giá trị / tổng giá trị



```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

- **sử dụng parameter lable để gán nhãn có các mục trong biểu đồ**

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

- vị trí bắt đầu mặc định sẽ đi từ trục x, vẽ ngược theo chiều kim đồng hồ
- **sử dụng parameter startangle để thay đổi vị trí bắt đầu**

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

- **explode: giúp xác định vị trí các mục so với trung tâm ( mặc định là null thì tất cả sát nhau)**
- **shadow: giúp thêm bóng vào biểu đồ**
- **color: dùng thay đổi màu các mục**

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow= True)
plt.show()
```

- **.legend() :tạo bảng chú giải**
- **parameter title :tạo tiêu đề chú giải**

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show()
```

# 8, Matplotlib Heatmap

- **Heatmap là một biểu đồ thống kê màu sắc**
-  **sử dụng để hiển thị mức độ tập trung hoặc phân phối của dữ liệu trong một bảng hoặc một không gian hai chiều.**
- **Trong heatmap, mỗi giá trị của bảng dữ liệu được biểu diễn bằng một ô màu có độ sáng hoặc màu sắc khác nhau, phản ánh giá trị của nó.**

```python
# library
import seaborn as sns
import pandas as pd
import numpy as np

# Create a dataset
df = pd.DataFrame(np.random.random((5,5)), columns=["a","b","c","d","e"])

# Default heatmap
p1 = sns.heatmap(df)
```

**Một số parameter**

sns.heatmap(data, *, vmin=None, vmax=None, cmap=None, center=None, robust=False, annot=None, fmt='.2g', annot_kws=None, linewidths=0, linecolor='white', cbar=True, cbar_kws=None, cbar_ax=None, square=False, xticklabels='auto', yticklabels='auto', mask=None, ax=None, **kwargs)



```python
import seaborn as sns
import numpy as np

data = np.random.rand(10, 12)
sns.heatmap(data)
```

cmap (Bảng màu): Tên hoặc đối tượng colormap của Matplotlib, hoặc danh sách các màu sắc. Nếu không được cung cấp, màu sắc mặc định sẽ phụ thuộc vào việc có đặt center hay không.

```python
sns.heatmap(data, cmap="YlGnBu")
```

**annot**: Nếu True, hiển thị giá trị dữ liệu trong mỗi ô. Nếu là một dạng array có cùng hình dạng với dữ liệu, sẽ sử dụng để chú thích heatmap thay vì giá trị dữ liệu.

```python
sns.heatmap(data, annot=True)
```

**fmt** (Định dạng Hiển Thị):giúp định dạng giới hạn của giá trị hiển thị trên mỗi ô.

```python
sns.heatmap(data, annot=True, fmt=".2f")
```

**linewidths **và **linecolor** (Độ Dày và Màu Viền): Xác định độ dày và màu của các đường biên giữa các ô.

```python
sns.heatmap(data, linewidths=0.5, linecolor="black")
```

**italicized textxticklabels** và **yticklabels** (Hiển Thị Nhãn Trục X và Y): Chọn liệu có hiển thị nhãn trục X và Y hay không.Cách hiển thị nhãn trục x và y. 'auto' tự động chọn.

```python
sns.heatmap(data, xticklabels=False, yticklabels=False)
```

**cbar** (Hiển Thị Than Chuột): Nếu True, vẽ thanh màu sắc.

```python
sns.heatmap(data,cbar=False)
```

**square** (Hiển Thị Ô Vuông):Nếu True, đặt tỷ lệ khía cạnh của Axes thành "equal" để mỗi ô sẽ có hình vuông.

```python
# sns.heatmap(data, square=True)
sns.heatmap(data, square=False)
```

**Một số parameter khác**


- **vmin, vmax**: Giá trị tối thiểu và tối đa để giới hạn mức màu sắc. Nếu không được cung cấp, giá trị này sẽ được suy luận từ dữ liệu và các đối số khác.



- **center**: Giá trị màu trung tâm khi vẽ biểu đồ heatmap cho dữ liệu lưỡng cảnh. Sử dụng tham số này sẽ thay đổi colormap mặc định nếu không có colormap nào được chỉ định.

- **robust**: Nếu True và không có vmin hoặc vmax, phạm vi màu sắc sẽ được tính toán với phân vị mạnh mẽ thay vì giá trị cực đại và cực tiểu.

- **annot**: Nếu True, hiển thị giá trị dữ liệu trong mỗi ô. Nếu là một dạng array có cùng hình dạng với dữ liệu, sẽ sử dụng để chú thích heatmap thay vì giá trị dữ liệu.


- **annot_kws**: Đối số dạng từ điển của các cặp khóa-giá trị được truyền vào matplotlib.axes.Axes.text khi annot là True.


- **cbar**: Nếu True, vẽ thanh màu sắc.

- **cbar_kws**: Đối số dạng từ điển của các cặp khóa-giá trị được truyền vào matplotlib.figure.Figure.colorbar.

- **cbar_ax**: Trục Axes của Matplotlib để vẽ thanh màu sắc, nếu không được cung cấp, sẽ sử dụng không gian từ Axes chính.

- **mask**: Mảng hoặc DataFrame kiểu bool, nếu có, dữ liệu sẽ không được hiển thị ở các ô tương ứng với True trong mask.

- **ax**: Trục Axes của Matplotlib để vẽ biểu đồ, nếu không được cung cấp, sẽ sử dụng Axes đang hoạt động.

**kwargs: Các tham số khác được chuyển vào matplotlib.axes.Axes.pcolormesh.

+ Được sử dụng để truy cập từng subplot trong một figure
+ matplotlib.pyplot.subplot(nrows=None, ncols=None, index=None, **kwargs)
+ nrows: số hàng của biểu đồ
+ ncols: số cột của biểu đồ
+ index: vị trí của biểu đồ con

```python
import matplotlib.pyplot as plt
import numpy as np

# Dữ liệu mẫu
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
z = np.array([7, 8, 9])

# Tạo subplot 3D
plt.subplot(1, 1, 1, projection='3d')

# Vẽ đồ thị 3D
plt.plot(x, y, z)

plt.show()
```

+ polar: Hệ trục polar là một loại hệ trục đặc biệt thường được sử dụng để vẽ đồ thị trong không gian hai chiều dựa trên các góc và bán kính thay vì các giá trị x và y

```python
import matplotlib.pyplot as plt
import numpy as np

# Dữ liệu mẫu
theta = np.linspace(0, 2*np.pi, 100)
r = 3 * np.sin(3 * theta)

# Tạo subplot với hệ trục polar
plt.subplot(1, 1, 1, polar=True)

# Vẽ đồ thị polar
plt.plot(theta, r)

plt.show()
```

+ aspect

```python
import matplotlib.pyplot as plt
import numpy as np

# Dữ liệu mẫu
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Tạo subplot với tỷ lệ "equal"
plt.subplot(1, 2, 1, aspect='equal')
plt.plot(x, y)
plt.title('Aspect "equal"')

# Tạo subplot với tỷ lệ tự động
plt.subplot(1, 2, 2, aspect='auto')
plt.plot(x, y)
plt.title('Aspect "auto"')

plt.show()
```

# 13, Biểu đồ phân tán 3 chiều


##Khái niệm

-Biểu đồ phân tán ba chiều là một công cụ trực quan

-Được sử dụng để hiển thị sự phụ thuộc hoặc mối quan hệ giữa ba biến hoặc nhóm dữ liệu trong không gian ba chiều.

-Nó là một phiên bản mở rộng của biểu đồ phân tán hai chiều, nơi các điểm dữ liệu được biểu diễn bằng các điểm trong không gian ba chiều

## Tác dụng

-Hiển thị mối quan hệ giữa ba biến

-Phát hiện mẫu và xu hướng

-Trực quan hóa dữ liệu lớn

-Hỗ trợ trong quyết định

```python
import numpy as np
import matplotlib.pyplot as plt

# Dữ liệu mẫu
height = [170, 165, 180, 155, 175]
weight = [65, 70, 75, 60, 80]
bmi = [22.49, 25.71, 23.15, 25.11, 26.12]

# Màu sắc tương ứng cho từng biến
colors = ['red', 'green', 'blue', 'yellow', 'purple']

# Vẽ biểu đồ phân tán 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(height, weight, bmi, c=colors)

# Đặt nhãn cho các trục
ax.set_xlabel('Chiều cao (cm)')
ax.set_ylabel('Cân nặng (kg)')
ax.set_zlabel('BMI')

# Hiển thị biểu đồ
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo dữ liệu mẫu với 100 điểm
np.random.seed(42)
n_points = 100
x = np.random.randint(140, 190, size=n_points)
y = np.random.randint(30, 80, size=n_points)
z = np.random.randint(20, 50, size=n_points)

# Tạo màu sắc tương ứng cho từng trục
x_colors = x
y_colors = y
z_colors = z

# Vẽ biểu đồ phân tán 3D với màu sắc từng trục
##tạo ra một đối tượng hình ảnh (figure) mới.
fig = plt.figure()
##chỉ số 111 cho thấy chúng ta đang tạo một hình ảnh với một trục duy nhất và không chia thành lưới các trục.
##Projection='3d' chỉ định rằng trục được tạo ra là một trục 3D để vẽ biểu đồ phân tán 3D.
ax = fig.add_subplot(111, projection='3d')

ax.scatter3D(x, y, z, c=x_colors, cmap='Accent_r', label='Chiều cao (cm)')
ax.scatter3D(x, y, z, c=y_colors, cmap='Greens', label='Cân nặng (kg)')
ax.scatter3D(x, y, z, c=z_colors, cmap='Blues', label='Tuổi (tuổi)')

# Đặt nhãn cho các trục
ax.set_xlabel('Chiều cao (cm)')
ax.set_ylabel('Cân nặng (kg)')
ax.set_zlabel('Tuổi (tuổi)')

# Hiển thị biểu đồ
plt.legend()
plt.show()
```

# 14, Hàm grid

```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo dữ liệu mẫu

 # Tạo ra một mảng x gồm 100 phần tử, bắt đầu từ 0 và kết thúc là 10, các phần tử trong mảng này được phân bố đều nhau trong khoảng từ 0 đến 10.
x = np.linspace(0, 10, 100)
 # Tạo ham sin(x)
y = np.sin(x)

# Tạo biểu đồ
plt.plot(x, y)

# Bật lưới trên trục x và y
plt.grid(True)

# Hiển thị biểu đồ
plt.show()
```

```python
matplotlib.pyplot.grid(b=None, which='major', axis='both', **kwargs)
```



b: Giá trị boolean hoặc None để xác định xem đường lưới có được bật hay tắt. Nếu True, đường lưới sẽ được bật. Nếu False, đường lưới sẽ được tắt. Nếu None (mặc định), đường lưới sẽ được bật nếu các định dạng đường lưới cụ thể được thiết lập.

---



which: Xác định đường lưới nào được vẽ. Có thể là 'major' (đường lưới chính) hoặc 'minor' (đường lưới phụ). Mặc định là 'major'.

---



axis: Xác định các trục nào sẽ có đường lưới. Có thể là 'both' (cả hai trục), 'x' (trục x) hoặc 'y' (trục y). Mặc định là 'both'.

---



**kwargs: Các đối số bổ sung được chuyển đến hàm matplotlib.lines.Line2D để tùy chỉnh định dạng đường lưới.


1.  Định dạng mặc định: Đường lưới có các đường dọc và ngang màu xám nhạt hoặc màu xám tối
2.  Màu sắc: Bạn có thể thiết lập màu sắc cho đường lưới bằng cách sử dụng đối số color.     
Ví dụ: plt.grid(True, color='red') sẽ thiết lập màu đỏ cho đường lưới.
3.Đường kẻ: Bạn có thể thiết lập kiểu và độ dày của đường kẻ trên đường lưới bằng cách sử dụng các đối số linestyle và linewidth.              
Ví dụ: plt.grid(True, linestyle='dashed', linewidth=0.5) sẽ thiết lập đường lưới với kiểu đứt gạch và độ dày 0.5.
4.Định dạng đường lưới cho trục x hoặc y: Bạn có thể chỉ định định dạng đường lưới cho trục x hoặc y bằng cách sử dụng đối số which.          
Ví dụ: plt.grid(True, which='major', axis='x') sẽ chỉ vẽ đường lưới chính trên trục x.
